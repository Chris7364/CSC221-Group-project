```python
from urllib.request import urlopen
from bs4 import BeautifulSoup
import re
import csv
import pandas as pd

url = 'https://en.wikipedia.org/wiki/List_of_highest-grossing_science_fiction_films'
page = urlopen(url)
soup2 = BeautifulSoup(page, 'html.parser')
print(soup2.prettify())
```

```python
# Find the Highest-grossing sci-fi films table on the page ... search for 'wikitable_sortable'
# soup2.find('table', class_ = 'wikitable sortable jquery-tablesorter')
all_sortable_tables = soup2.find_all('table', class_ = 'wikitable sortable')
print("HOW MANY TABLES DID I FIND? -- " + str(len(all_sortable_tables)))
for table in all_sortable_tables:
    print("Table:" + str(table) + "/n")
```

```python
# Find the Highest-grossing sci-fi films table ... this is the first - index [0]
right_table = all_sortable_tables[0]
print("Table:" + str(right_table) + "/n")
```

```python
# Take a first-look at the ROWS -- <td> TAGS 
# How many cells (columns) are there on each ROW? 
for row in right_table:
    cells = right_table.find_all("td")
    print("\n Length of this cell is : ", len(cells))
    print("\n Its Cell values ")
    for cell in cells:
      print("\n", cell)

    
```

```python
# Use the information gathered above to pick the cells (columns) of data I want 
A = []
B = []
C = []
D = []

for row in right_table.findAll('tr'):
    cells=row.findAll('td')
    if len(cells)==5:
        A.append(cells[0].find(text=True))
        B.append(cells[1].find(text=True))
        C.append(cells[2].find(text=True))
        D.append(cells[3].find(text=True))
        
    
```

```python
import pandas as pd
df = pd.DataFrame({'Rank':A, 'Film':B, 'Worlwide Gross':C, 'Year':D})
print(df)
```
